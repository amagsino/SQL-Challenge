```python
from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from password import password
from db_name import db
```

```python
engine = create_engine(f'postgres://postgres:{password}@localhost:5432/{db}')
connection = engine.connect()
```

```python
# Display employee number, salary, from and to date from salaries
salaries_data = pd.read_sql("SELECT * FROM salaries", connection)
salaries_data.head()
```

```python
# Display employee number, title, from and to date from titles
titles_data = pd.read_sql("SELECT * FROM titles", connection)
titles_data.head()
```

```python
# Combine both salaries and titles databases
combined_data = pd.merge(salaries_data, titles_data, on="emp_no", how="inner")
combined_data.head()
```

```python
# Group data by title
grouped_df = combined_data.groupby("title").mean()
grouped_df
```

```python
# Drop employee number column
revised_grouped_df = grouped_df.drop(columns = "emp_no")
revised_grouped_df
```

```python
# Reset index
revised_grouped_df = revised_grouped_df.reset_index()
revised_grouped_df
```

```python
# Set x_axis, y_axis & Tick Locations
x_axis = revised_grouped_df["title"]
ticks = np.arange(len(x_axis))
y_axis = revised_grouped_df["salary"]
 
# Create Bar Chart Based on Above Data
plt.bar(x_axis, y_axis, align="center", alpha=0.5, color=["y", "b", "m", "g", "c", "r", "k"])

# Create Ticks for Bar Chart's x_axis
plt.xticks(ticks, x_axis, rotation="vertical")

# Set Labels & Title
plt.ylabel("Salaries ($)")
plt.xlabel("Employee Titles")
plt.title("Average Employee Salary by Title")

# Show plot
plt.show()
```
